```python
from PIL import Image
import numpy as np
```

```python
im = Image.open('taiwan.png')
```

```python
arr = np.array(im.convert('L').resize((90,38)))
```

```python
s = ""
for i in range(1,arr.shape[0]-3):
    s += "".join("0" if x else "1" for x in arr[i][24:-24]>5)
    print("".join("0" if x else "1" for x in arr[i][24:-24]>5))
```

```python
import gmpy2
```

```python
n = int(s, 2)
gmpy2.is_prime(n)
```

```python
for i in range(1,len(s)):
    p = n^(1<<i)
    if gmpy2.is_prime(p):
        print(p)
        break
```

```python
s2 = bin(p)[2:]
s2 = '0'*(len(s)-len(s2))+s2
s2
```

```python
n = int(s2, 2)
gmpy2.is_prime(n)
```

```python
for i in range(0,len(s2),42):
    print(s2[i:i+42])
```
