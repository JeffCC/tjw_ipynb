```python
if "map_mem" not in globals():
    map_mem=dict()
w=widgets.HTMLWidget()
w2=widgets.HTMLWidget()
def show_maze2():
    html ="<table>"
    for i in range(-44,45):
        html+="<tr>"
        for j in range(-44,45):
            mx,my = j, i
            bg = {'@': '#000', '.': '#fff',  '':'#0f7' }[map_mem.get((mx,my), '')]
            html+='<td style="background-color: %s; width: 6px; height: 6px;">%c</td>'%(bg, ' ')
        html+="</tr>"
    html+="</table>"
    w2.value = html
    
def show_maze(d):    
    html ="<table>"
    for i in range(9):
        html+="<tr>"
        for j in range(9):
            try:
                mx,my = cx+j-4,cy+i-4
                bg = {'@': '#000', '.': '#fff', 'o':'#777' }[d[i][j]]
                if (mx,my) in dead:
                    bg="#f00"
                elif (mx,my) in old:
                    bg="#0f0"                
            except:
                bg = "#00f"
            t = d[i][j]
            if t=='o': t='.'
            if t in "@.":
                if (mx,my) in map_mem:
                    if map_mem[mx,my] != t:
                        raise xxxxxx
                else:
                    map_mem[mx,my] = t
            html+='<td style="background-color: %s;width: 18px; height: 18px;">%c</td>'%(bg, d[i][j])
        html+="</tr>"
    html+="</table>"
    w.value = html
```
