```python
var('x', real=True)
var('a', real=True)
# a = Symbol('x', real=True, positive=True)
def phi(n):
    return sqrt(2/a)*sin(n*pi*x/a)
formula = integrate(phi(1)*phi(1),(x,0,a))*a/2
formula
```

```python
refine(formula, Q.is_true(a>0))
```
